# Query ODE/PDS product-types

ODE provides a REST interface to query about (image) data products available in PDS data archives.
ODE interface provide a series of parameters to search for products, we will use a bounding-box to define a location.

The quest for products through ODE API is implemented in the GPT library, check the source code there if you feel like getting into the technical details/workflow.

Relevant topics:
- ODE REST API: http://oderest.rsl.wustl.edu/
    * API manual: http://oderest.rsl.wustl.edu/ODE_REST_V2.1.pdf


```python
import gpt

from gpt.search import ode
```

```python
dsets = ode.available_datasets('mars')

dsets
```

```python
dsets.loc['Mars','MRO']
```

```python
from importlib import reload
from pathlib import Path
```

```python
# Define the region to search for products/footprints
#
bounding_box = [-0.5, -0.5, 1.5, 1.5]
```

## First, use the functions available in GPT-search' ODE module

```python
results = ode.search(bounding_box, target='mars', ihid='mro', iid='ctx', pt='edr', match='contain')

#print(results_ctx)
```

```python
products = ode.parse_products(results, 
                                  data_selectors={'Type':['product','browse'], 'FileName': ['JPEG$','IMG$']}, 
                                  data_select_how='all')
products[:max(2,1)]
```

```python
gdf = ode.to_geodataframe(products, geometry_field='Footprint_C0_geometry')
gdf
```

```python
%matplotlib inline

gdf.plot(alpha=0.25, edgecolor='red', legend=True, column='pdsid', legend_kwds={'loc': 'upper left', 'bbox_to_anchor': (1, 1)}, figsize=(12,8))
```

```python
del gdf['notes']

gdf.to_file('search_ctx.geojson', driver='GeoJSON')
```

```python
ihid='mex'
iid='hrsc'
pt='refdr3'

bounding_box = [-0.5, -0.5, 0.5, 0.5]
```

```python
results = ode.search(bounding_box, target='mars', ihid=ihid, iid=iid, pt=pt, match='intersect')

products = ode.parse_products(results, 
                                  data_selectors={'Type':'product', 'FileName': '_ND'}, 
                                  data_select_how='all')

# products = products[::4]

gdf = ode.to_geodataframe(products, geometry_field='Footprint_C0_geometry')
```
